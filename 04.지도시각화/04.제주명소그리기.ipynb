# 제주명소 그리기

```python
import folium
import pandas as pd
```

```python
df = pd.read_csv('제주명소2.csv')
```

```python
df
```

```python
map = folium.Map(location=[df['위도'].mean(),df['경도'].mean()],zoom_start=11) # 가장기본적인 오픈스트리트 맵


for i in df.index:
    folium.Marker(
        location=[df['위도'][i],df['경도'][i]],
        tooltip=df.명소[i],
        popup=folium.Popup(df.도로명주소[i],max_width=200)  
    ).add_to(map)
title = title_html = '<h3 align="center" style="font-size:20px">제주 명소 10개</h3>' 
map.get_root().html.add_child(folium.Element(title_html))

map
```

```python
map.save('제주명소.html')

df['0'] = 0
```
