```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus'] = False
mpl.rcParams['font.family'] = 'Malgun Gothic'
```

```python
plt.plot([0,1,2,3],[0,1,4,9],'ms--', linewidth=3, markersize=10) #  linewidth 굵기 markersize  = 마커의 크기 
plt.plot([8,4,2,1],'g^:',lw=5,ms=20)
plt.title('Magenta-Square-Dashed,Green-삼각형- 점선')
plt.grid()
plt.show()
```

- 그림 범위

```python
plt.plot([0,1,2,3],[0,1,4,9],'ms--', linewidth=3, markersize=10) #  linewidth 굵기 markersize  = 마커의 크기 
plt.plot([8,4,2,1],'g^:',lw=5,ms=20)
plt.title('x축, y축 범위 지정')
plt.xlim(-1,4)
plt.ylim(-1,10)
plt.grid()
plt.show()
```

- Tick 설정

```python
X = np.linspace(-np.pi, np.pi, 500)
Y = np.cos(X)
plt.title('Tick 설정')
plt.plot(X,Y)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1,0,1])
plt.grid()
plt.show()
```

- 범례(legned) 

```python
X = np.linspace(-np.pi, np.pi, 500)
C, S = np.cos(X), np.sin(X)
plt.title('Legend를 표시한 플롯')
plt.plot(X,C, label='cosine')
plt.plot(X,S, label='Sine')
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1,0,1])
plt.grid()
plt.legend()
plt.show()
```

- x축 , Y축 label

```python
X = np.linspace(-np.pi, np.pi, 500)
C, S = np.cos(X), np.sin(X)
plt.title('Legend를 표시한 플롯')
plt.plot(X,C, label='cosine')
plt.plot(X,S, label='Sine')
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1,0,1])
plt.grid(),plt.legend()
plt.xlabel('time'),plt.ylabel('Amplitude')
plt.show()
```

- 그림의 크기

```python
plt.figure(figsize=(10,6))
plt.title('Legend를 표시한 플롯')
plt.plot(X,C, label='cosine')
plt.plot(X,S, label='Sine')
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1,0,1])
plt.grid(),plt.legend()
plt.xlabel('time'),plt.ylabel('Amplitude')
plt.show()
plt.savefig('삼각함수.png')
```
