```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus'] = False
mpl.rcParams['font.family'] = 'Malgun Gothic'
```

```python
np.random.seed(2021)
df = pd.DataFrame(np.random.randn(100,3),
    index=pd.date_range('1/1/2021',periods=100),
    columns= list('ABC')).cumsum()
df.plot()
plt.title('Pandas의 Plot method 사용 예')
plt.xlabel('일자')
plt.ylabel('Data')
plt.show()
```

### 2.Bar chart

```python
import seaborn as sns
iris = sns.load_dataset('iris')

plt.figure(figsize=(12,6))
iris.sepal_length[:50].plot(kind='bar', rot=0) # 그래프의 모양을 나타냄~~  # 봉봉스타일
plt.title('Setosa 품종의 Sepal length')
plt.show()
```

```python
df2 = iris.groupby('species').mean()
df2.columns.name = 'Feature'
df2
```

```python
df2.plot.bar(rot=0)
plt.title('iris 각 품종의 feature별 평균')
plt.ylim(0,9) # y 의 리미트를 늘려서 범례가 안가리게 만듬 
```

```python
df2.plot.bar(rot=0)
plt.title('iris 각 품종의 feature별 평균')
plt.ylim()
```

# Pie Chart

```python
titanic = sns.load_dataset('titanic')
s = titanic['class'].value_counts()
s
```

```python
s.plot.pie(autopct='%2.f%%')
plt.title('선실별 승객수')
plt.show()
```

### 4.Box plot

```python
iris[iris.species=='setosa'].plot.box(figsize=(8,10))
plt.title('Setosa 품종의 각 Featur별 Box Plot')
plt.show()
```

```python
iris.boxplot(by='species',figsize=(10,8))
# plt.title('각 피쳐의 품종별 데이터에대한 box plot')
plt.show()
```
