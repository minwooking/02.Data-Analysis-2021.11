```python
X = np.random.randn(100) 
Y = np.random.randn(100)
plt.title('산점도')
plt.scatter(X,Y)
plt.grid()
plt.show()
```

```python
colors = np.random.rand(100) # 색상이 나오는 
area = np.pi * (15 * np.random.rand(100))**2
plt.scatter(X,Y, s=area, c=colors, alpha=0.5) # alph 투명도
plt.title('Bubble Chart')
plt.show()
```

# Box plot

```python
# Setosa 품종
iris = sns.load_dataset('iris')
setosa = iris[iris.species == 'setosa']
```

```python
plt.figure(figsize=(8,10))
plt.boxplot([setosa.sepal_length,setosa.sepal_width,setosa.petal_length,setosa.petal_width],labels=iris.columns[:-1])
plt.title('Setosa 품종의 각 피쳐별 박스 플롯',fontsize=20)
plt.show()
```

### 5. 파이(pie) 차트 

```python
ratio = [34,32,16,18]
labels = ['사과','바나나','멜론','포도']
plt.pie(ratio, labels=labels,autopct='%.1f%%')
plt.show()
```
