# Data Frame -합성

```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df1 = pd.DataFrame({
'고객번호': [1001, 1002, 1003, 1004, 1005, 1006, 1007],
'이름': ['둘리', '도우너', '또치', '길동', '희동', '마이콜', '영희'] }, columns=['고객번호', '이름'])
df1
```

```python
df2 = pd.DataFrame({
'고객번호': [1001, 1001, 1005, 1006, 1008, 1001], '금액': [10000, 20000, 15000, 5000, 100000, 30000]
}, columns=['고객번호', '금액'])
df2
```

```python
df3 = pd.merge(df1,df2) # 교집합만 출력
df3
```

```python
# full outer join
df3 = pd.merge(df1,df2,how='outer') # 어떻게 교집합이 아닌것도 같이 출력
df3
```

```python
# left outer join
df3 = pd.merge(df1,df2,how='left') # 어떻게 교집합이 아닌것도 같이 출력
df3
```

```python
df2 = pd.DataFrame({
'고객번호': [1001, 1001, 1005, 1006, 1008, 1001], '금액': [10000, 20000, 15000, 5000, 100000, 30000]
}, columns=['고객번호', '금액'])
df2
```

```python
# Right Outer join
df3 = pd.merge(df1,df2,how='right') # 어떻게 교집합이 아닌것도 같이 출력
df3
```

```python
df3 = df1.set_index('고객번호')
df4 = df2.set_index('고객번호') # 인덱스를 지정해주고 동일인덱스를 기준으로 만듬 
df3.join(df4)
```

```python
pop = pd.read_csv('서울시자치구별주민등록인구.txt',sep='\t',skiprows=2)
pop.head()
pop.count()
```

```python
# 자치구 계 만 남겨두고 다없애버리기
# 자치구를 인덱스
# 계를 총인구
# 노인인구를 65세이상 고령자로 수정하지 

pop1 = pop[['자치구']]+pop[['65세이상고령자']]+pop[['계']]
pop1.count()
```

```python
# 필요한 컬럼만 가져오기 
pop = pop[['자치구','계','65세이상고령자']]
pop.head(3)
```

```python
# 2) 합계 지우기
pop.drop([0],inplace=True)
pop.head(3)
```

```python
# 3) 자치구를 인덱스로, 컬럼명 변경
# pop.set_index('자치구',inplace=True)
pop.rename(columns={'계':'총인구','65세이상고령자':'노인인구'},inplace=True)
pop.head(3)
```

```python
# 4) 숫자를 정수로 변환
pop = pop.apply(lambda x: x.str.replace(',','')).astype(int)
pop.head(3)
```

```python
# 5) 노인인구비율
pop['노인인구비율'] = (pop['노인인구']/pop['총인구']*100).round(2)
pop.head()
```

# cctv data 읽어서 인구 데이터와 합치기

```python
cctv = pd.read_csv('cctv.csv')
cctv.head(3)
```

```python
cctv.set_index('자치구',inplace=True)
```

```python
cctv.head()
```

```python
df= pop.join(cctv)
df.head()
```

```python
# 인구 1천명당 CCTV 댓수
df['1천명당 CCTV댓수'] = (df.CCTV댓수/df.총인구 * 1000).round(2)
df.tail()
```

```python
df.to_csv('서울시자치구별인구와CCTV.csv')
```
